## Splitting into train-test using user input ratio

```
def split(df,splitratio):
  size=int(len(df)*splitratio)
  train=df.iloc[0:size]
  test=df.iloc[size:]
  return (train,test)
```

```
train,test=split(df,0.8)
```

## Function to TRAVERSE row and columns


```
#for traversing each columns
def traverse_columns(df):
  for i in df.columns:
    print(i)
```

```
#for traversing each rows
def traverse_rows(df):
  for i in df.index:
    print(df.loc[i,:])
    print('end of row '+str(i))
```

## Write a function to query if a values exists in the dataframe and return its complete column and complete row separately

```
def searchvalue(df,value):
  for i in range(0,len(df.columns)):
    if value in list(df.iloc[:,i]):
      column=df.iloc[:,i]
      for j in range(0,len(column)):
        if value==column[j]:
          row=df[df.index==j]
          return(row,column)
```

```
element_row,element_column=searchvalue(df,-114.31)
```

## From the list of columns , construct all combinations of 3 values and for each combination combine their row values as  a combined column

```
def column_combination(df):
  combined_list=[]
  append_dataframe=[]
  columns=df.columns
  for i in range(0,len(columns)-2):
    combined_list.append((columns[i],columns[i+1],columns[i+2]))
  for i in combined_list:
    dataframe=pd.DataFrame()
    for j in i:
      dataframe[j]=df.loc[:,j]
    append_dataframe.append(dataframe)
  return append_dataframe
```

```
combination_dataframe=column_combination(df)
```
